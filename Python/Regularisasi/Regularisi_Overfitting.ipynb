## Buat fungsi sigmoid dan fitur map

```python
n, m = Data.shape
```

```python
def sigmoid(z):
    return 1.0 / (1.0 + np.exp(-z))
```

```python
def map_feature(x1,x2):
    """
    Dalam kasus data seperti ini kita perlu polynomial hipotesis
    bukan hanya linear output dari map_feature adalah polynomial
    x1,x, x1^2, x2^2, x1*x2, x1*x2^2, dan seterusnya
    """
    degree = 7
    out = [np.ones_like(x1)]
    for ii in np.arange(1,degree):
        for jj in range(ii + 1):
            
            out.append(x1**(ii-jj)*x2**jj)
            
    return np.array(out)
```

```python
def cost_function_reg(theta, X, y, Lambda):
    """
    Parameter
    ---------
    theta: estimasi data
    X: Data sebaran 
    y: Label
    Lambda: nilai regularisasi
    
    Return
    ------
    J: Cost function
    grad: gradient
    """
    # Inisial banyaknya training
    m = len(y) 
    
    J = 0
    grad = np.zeros_like(theta)
 
    # Sigmoid
    hyp = np.squeeze(sigmoid(theta@X))
    err = (hyp - y)

    grad = ((X@err)/m) + (Lambda* theta)/m
    grad[0] -=  (Lambda * theta[0])/m
    
    # Regularisasi
    reg = (Lambda / (2*m)) * np.sum(theta[1::]**2)
     
    J = (1/m)*((-y @ np.log(hyp) - (1 - y)@np.log(1 - hyp))) + reg
     
    return J, np.squeeze(grad)
```

## Hitung dan update gradient

## Gunakan fungsi scipy

```python
import scipy.optimize
args = (Xpol, y, Lambda)
initial_theta = np.zeros((Xpol.shape[0]))
# Optimize our weights using the Conjugate Gradient algorithm
final_theta = scipy.optimize.minimize(cost_function_reg, initial_theta, method = 'CG', args = args,
                       jac=True, options = {'maxiter':50})
```

```python
u = np.linspace(-1, 2, 50) 
v = np.linspace(-1, 2, 50)

z = np.zeros((len(u), len(v)))

for ii in range(len(u)):
    for jj in range(len(v)):
        z[ii,jj] = map_feature(u[ii], v[jj])@final_theta.x
    
```
