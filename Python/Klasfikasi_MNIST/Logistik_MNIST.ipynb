```python
import numpy as np
import scipy.io
from math import sqrt
import scipy.optimize
import matplotlib.pyplot as plt
```

```python


# Import data MNIST
mat = scipy.io.loadmat('mnist.mat')

# Import data training
data_X = mat['trainX']
data_y = mat['trainY'].T

# Normalisasi data
data_X = data_X/255.0

# Bagi training dan validasi data dengan sekitar 50000 untuk training dan 
# sisanya sekitar 10000 untuk validasi

sample = 50000
X_train = data_X[:sample, : ]

X_val = data_X[sample:, : ]

y_train = data_y[:sample, : ]

y_val = data_y[sample:, : ]

# Import test set
X_test = mat['testX']
y_test = mat['testY'].T

# Normalisasi test set
X_test = X_test/255.0
```

```python
# Fungsi untuk menghitung sigmoid
def sigmoid(z):
    """" 
    Hitung fungsi sigmoid pada input z, bisa skalar, vektor atau matriks
    """
    ans = 1/(1+np.exp(-z))
    return ans
```

```python
def cost_function_reg(theta, X, y, Lambda):
    """
    Fungsi loss pada regresi
    theta: inisiasi
    X: data
    y: label
    Lambda: regularisasi
    
    Return
    ------
    J: Nilai loss
    grad: gradient
    """
    # Inisial banyaknya training
    m = len(y) 
    
    
    grad = np.zeros_like(theta)
 
    # Sigmoid 
    hyp = np.squeeze(sigmoid(theta@X))
    err = (hyp - y)

    grad = ((X@err)/m) + (Lambda* theta)/m
    grad[0] -=  (Lambda * theta[0])/m
    
    # Regularisasi
    reg = (Lambda / (2*m)) * np.sum(theta[1::]**2)
     
    J = (1/m)*((-y @ np.log(hyp) - (1 - y)@np.log(1 - hyp))) + reg
    print('Cost function ', J)
    return J, np.squeeze(grad)
```

```python
def one_vs_all(X, y, num_labels, Lambda):
    """
    Fungsi satu versus semua pada klasifikasi
    num_labels:banyaknya label
    X: data
    y: label
    Lambda: regularisasi
    
    Return
    ------
    all_theta: nilai optimal
    """
    m,n = X.shape 

    all_theta = np.zeros((num_labels, n + 1))
 
 
    X = np.hstack((np.ones((m,1)), X))
    X = X.T 
    initial_theta = np.zeros((1,n + 1))

    
    for c in range(num_labels):
        args = (X , (y == c).astype(int), Lambda) 
        res_theta = scipy.optimize.minimize(cost_function_reg, initial_theta, method = 'CG', args = args,
                       jac=True, options = {'maxiter':50})
        all_theta[c,:] = res_theta.x
        
    return all_theta
```

```python
def predict_one_vs_all(all_theta, X):
    """
    Fungsi untuk prediksi satu vs semua
    all_theta: nilai optimal
    X: Data
 
    
    Return
    ------
    p: index
  
    """
    m = X.shape[0]
    num_labels = all_theta.shape[0]
    
    p = np.zeros((m))
    
    X = np.hstack((np.ones((m,1)), X))
    
    X = X.T
    
    new_sig = sigmoid(all_theta @ X)

    for ii in range(m):
        
        index = np.argmax(new_sig[:,ii])
        p[ii] = index
    return p
```

```python

theta_t = np.array([-2, -1, 1, 2])
X_t = np.zeros((4,5))
X_t[0,:] = np.ones(5)
X_t[1::,:] = np.arange(1,16).reshape(3,5)/10
y_t = np.array([1,0,1,0,1])
lambda_t = 3;
J, grad = cost_function_reg(theta_t, X_t, y_t, lambda_t);
```

```python
Lambda = 0.1
num_labels = 10
```

```python
all_theta = one_vs_all(X_train, np.squeeze(y_train), num_labels, Lambda)
```

```python
pred_train = predict_one_vs_all(all_theta, X_train)
accu_train = np.mean(pred_train==np.squeeze(y_train))*100
```

```python
pred_test = predict_one_vs_all(all_theta, X_test)
accu_test = np.mean(pred_test==np.squeeze(y_test))*100
```

```python
pred_val = predict_one_vs_all(all_theta, X_val)
accu_val = np.mean(pred_val==np.squeeze(y_val))*100
```
